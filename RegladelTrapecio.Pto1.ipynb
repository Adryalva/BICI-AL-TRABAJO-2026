## Regla del trapecio

Importar librerias

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, lambdify, sympify
import matplotlib.cm as cm
```

Definir función a integrar, límites de integración y número de subintervalos

```python
# Solicitar la función y los límites de integración
funcion_str = input("Introduce la función a integrar : ")
a = float(input("Introduce el límite inferior de integración: "))
b = float(input("Introduce el límite superior de integración: "))
n = int(input("Introduce el número de intervalos (trapecios) que deseas utilizar: "))

# Convertir la función de cadena a una expresión simbólica y luego a una función lambda
x = symbols('x')
funcion_expr = sympify(funcion_str)
funcion = lambdify(x, funcion_expr)
```

Implementar la regla del Trapecio

```python
#Calcular la integral usando la regla del trapecio
x_vals = np.linspace(a, b, n+1)
y_vals = funcion(x_vals)
h = (b - a) / n

# Calcular la integral usando la fórmula de la regla del trapecio
integral_trapecio = (h/2) * (y_vals[0] + 2*sum(y_vals[1:-1]) + y_vals[-1])

# Mostrar el resultado de la integral aproximada
print(f"La integral aproximada usando la regla del trapecio es: {integral_trapecio}")
```

Realizar la gráfica

```python
#Graficar la función y los trapecios
# Crear puntos para una gráfica más suave de la función
x_fine = np.linspace(a, b, 400)
y_fine = funcion(x_fine)

plt.figure(figsize=(10, 6))

# Graficar la función
plt.plot(x_fine, y_fine, 'b', label=f"f(x) = {funcion_str}")
plt.fill_between(x_fine, y_fine, alpha=0.2, label="Área bajo la curva")

# Seleccionar una paleta de colores para los trapecios
colors = cm.viridis(np.linspace(0, 1, n))

# Graficar los trapecios, cada uno con un color diferente
for i in range(n):
    plt.fill_between([x_vals[i], x_vals[i+1]], [y_vals[i], y_vals[i+1]], color=colors[i], alpha=0.7)

# Configuración de la gráfica
plt.title("Aproximación de la integral usando la regla del trapecio")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x)", "Área bajo la curva"])
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
